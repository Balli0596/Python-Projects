```python
import os

IMG_DIR = "/content/chg1_pipeline/images"
PREP_DIR = "/content/chg1_pipeline/preprocessed"

os.makedirs(PREP_DIR, exist_ok=True)

print("Input images:", len(os.listdir(IMG_DIR)))
print("Preprocessed dir ready:", PREP_DIR)
```

```python
import cv2
import numpy as np

def preprocess_for_ocr(img):
    """
    OCR-friendly preprocessing
    """
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Mild denoising
    denoised = cv2.fastNlMeansDenoising(gray, h=15)

    # Adaptive Gaussian threshold (good for text)
    thresh = cv2.adaptiveThreshold(
        denoised,
        255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY,
        31,
        10
    )
    return thresh


def preprocess_for_radio(img):
    """
    Radio-button-friendly preprocessing
    """
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Light blur to smooth noise
    blur = cv2.GaussianBlur(gray, (5, 5), 0)

    # Keep grayscale (do NOT threshold aggressively)
    return blur
```

```python
for fname in sorted(os.listdir(IMG_DIR)):
    if not fname.endswith(".png"):
        continue

    img_path = os.path.join(IMG_DIR, fname)
    img = cv2.imread(img_path)

    ocr_img = preprocess_for_ocr(img)
    radio_img = preprocess_for_radio(img)

    base = fname.replace(".png", "")

    ocr_path = os.path.join(PREP_DIR, f"{base}_ocr.png")
    radio_path = os.path.join(PREP_DIR, f"{base}_radio.png")

    cv2.imwrite(ocr_path, ocr_img)
    cv2.imwrite(radio_path, radio_img)

    print("Processed:", fname)
```

```python
tokens = []
boxes = []

for i in range(len(ocr_data["text"])):
    word = ocr_data["text"][i].strip()
    if word == "":
        continue

    x = ocr_data["left"][i]
    y = ocr_data["top"][i]
    w = ocr_data["width"][i]
    h = ocr_data["height"][i]

    # Normalize box to 0–1000 scale
    x0 = int(1000 * x / width)
    y0 = int(1000 * y / height)
    x1 = int(1000 * (x + w) / width)
    y1 = int(1000 * (y + h) / height)

    tokens.append(word)
    boxes.append([x0, y0, x1, y1])
```

```python
print("Total tokens:", len(tokens))
print("Total boxes:", len(boxes))

assert len(tokens) == len(boxes), "❌ Tokens and boxes mismatch!"
```

```python
from transformers import LayoutLMv3Processor, LayoutLMv3ForTokenClassification
from PIL import Image
import torch

# Processor
processor = LayoutLMv3Processor.from_pretrained(
    "microsoft/layoutlmv3-base",
    apply_ocr=False
)

# Trained model
model = LayoutLMv3ForTokenClassification.from_pretrained(
    "/content/layoutlmv3_chg1/checkpoint-45"
)

model.eval()
```

```python
# Example: testing on page 1
image_path = "/content/chg1_pipeline/preprocessed/page_1_ocr.png"
tokens_test = tokens          # tokens of page 1
boxes_test = boxes            # boxes of page 1

image = Image.open(image_path).convert("RGB")
```

```python
encoding = processor(
    image,
    tokens_test,
    boxes=boxes_test,
    word_labels=None,   # No labels needed during inference
    padding="max_length",
    truncation=True,
    return_tensors="pt"
)

# LayoutLMv3 expects 3D/4D input
pixel_values = encoding["pixel_values"]  # shape: [1, 3, 224, 224]
input_ids = encoding["input_ids"]
bbox = encoding["bbox"]
attention_mask = encoding["attention_mask"]
```

```python
with torch.no_grad():
    outputs = model(
        input_ids=input_ids,
        bbox=bbox,
        attention_mask=attention_mask,
        pixel_values=pixel_values
    )

logits = outputs.logits  # [batch_size, seq_len, num_labels]
predicted_ids = logits.argmax(-1).squeeze().tolist()
```

```python
# id2label from training
predicted_labels = [model.config.id2label[i] for i in predicted_ids]

# Show tokens with predicted labels
for token, label in zip(tokens_test, predicted_labels[:len(tokens_test)]):
    print(f"{token} -> {label}")
```
